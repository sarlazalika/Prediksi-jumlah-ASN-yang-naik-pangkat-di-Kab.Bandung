```python
import numpy as np
import matplotlib.pyplot as plt

# Data historis jumlah ASN yang naik pangkat
data_historis = [1802, 1584, 2039]
tahun_historis = [2021, 2022, 2023]  # Tahun terkait data historis

# Parameter simulasi
n_simulasi = 10000  # Jumlah iterasi Monte Carlo

# Menghitung rata-rata dan standar deviasi data historis
mean_data = np.mean(data_historis)
std_dev_data = np.std(data_historis)

# Simulasi Monte Carlo
simulasi = np.random.normal(mean_data, std_dev_data, n_simulasi)

# Menangkap nilai yang masuk dalam batas logis (jumlah ASN tidak bisa negatif)
simulasi = simulasi[simulasi > 0]

# Statistik hasil simulasi
rata_rata_simulasi = np.mean(simulasi)
std_dev_simulasi = np.std(simulasi)

# Hasil simulasi Monte Carlo
print(f"Hasil Simulasi Monte Carlo (Jumlah ASN yang Naik Pangkat):")
print(f"- Rata-rata: {rata_rata_simulasi:.2f}")
print(f"- Standar Deviasi: {std_dev_simulasi:.2f}")

# Visualisasi hasil simulasi
plt.figure(figsize=(10, 6))
plt.hist(simulasi, bins=50, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(rata_rata_simulasi, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {rata_rata_simulasi:.2f}')
plt.title('Distribusi Simulasi Monte Carlo (Jumlah ASN Naik Pangkat)')
plt.xlabel('Jumlah ASN yang Naik Pangkat')
plt.ylabel('Frekuensi')
plt.legend()
plt.grid(alpha=0.3)

# Menampilkan data historis sebagai referensi
for i, tahun in enumerate(tahun_historis):
    plt.annotate(f"{tahun}", xy=(data_historis[i], 50), xytext=(data_historis[i] + 50, 60),
                 arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=10)

plt.show()
```
